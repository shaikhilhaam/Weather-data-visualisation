```python
g1.rename(columns={'WindSpeedkm/h':'WindSpeed'},inplace=True)
plt.figure(figsize=(12,7))
plt.plot(g1.Date,g1.WindSpeed, label='wind(km/hr)', marker='o')
plt.plot(g1.Date,g1.TemperatureC, label='temp(C)',marker='o')
plt.plot(g1.Date,g1.Visibilitykm, label='visi(km)',marker='o')
plt.legend()
plt.xlabel('years')
plt.ylabel('values')
plt.xticks(g1.Date)
plt.title('Mean values in each year')
plt.show()
```

```python
g1.rename(columns={'WindSpeedkm/h':'WindSpeed'},inplace=True)
plt.figure(figsize=(8,5))
plt.scatter(g1.Date,g1.WindSpeed, label='wind(km/hr)', marker='o')
plt.scatter(g1.Date,g1.TemperatureC, label='temp(C)',marker='o')
plt.scatter(g1.Date,g1.Visibilitykm, label='visi(km)',marker='o')
plt.legend(loc='lower right')
plt.xlabel('years')
plt.ylabel('values')
plt.title('Mean values in each year')
plt.show()
```

```python
g1.rename(columns={'WindSpeedkm/h':'WindSpeed'},inplace=True)
plt.figure(figsize=(10,10))
plt.bar(g1.Date,g1.WindSpeed, label='wind(km/hr)',width=-20,align='edge')
plt.bar(g1.Date,g1.TemperatureC, label='temp(C)',width=20,align='edge')
plt.bar(g1.Date,g1.Visibilitykm, label='visi(km)',width=40,align='edge')
plt.legend(loc='lower right')
plt.xlabel('years')
plt.ylabel('values')
plt.title('Mean values in each year')
plt.show()
```
